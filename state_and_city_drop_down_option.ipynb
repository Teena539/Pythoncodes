```python
from google.colab import files
import os
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Upload the file if it's not already present
if not os.path.exists('/content/weather_dataset (1).csv'):
    uploaded = files.upload()
    os.listdir()

# Read the dataset
data = pd.read_csv('/content/weather_dataset (1).csv')

# Create state dropdown widget
state_dropdown = widgets.Dropdown(
    options=sorted(data['State'].unique()),
    description='Select State:'
)
# Create City dropdown (initially empty)
city_dropdown = widgets.Dropdown(
    options=[],
    description='Select City:'
)

# Output area for temperature and humidity
output = widgets.Output()

# Update city dropdown options when state changes
def update_city_options(change):
    selected_state = change.new
    city_dropdown.options = sorted(data[data['State'] == selected_state]['City'].unique())
    city_dropdown.value = None  # Reset city selection

# Display weather data based on city selection
def display_weather_data(change):
    with output:
        clear_output(wait=True)
        selected_city = change.new
        if selected_city:
            city_data = data[(data['State'] == state_dropdown.value) & (data['City'] == selected_city)]
            if not city_data.empty:
                temperature = city_data['Temperature (°C)'].iloc[0]
                humidity = city_data['Humidity (%)'].iloc[0]

                # Display temperature and humidity
                print(f"Temperature for {selected_city}: {temperature}")
                print(f"Humidity for {selected_city}: {humidity}")

                # Create bar chart
                plt.figure(figsize=(6, 4))
                plt.bar(['Temperature', 'Humidity'], [temperature, humidity],
                        color=['skyblue', 'lightcoral'])
                plt.ylabel('Value')
                plt.title(f'Temperature and Humidity for {selected_city}')
                plt.show()
            else:
                print(f"No data found for {selected_city}.")
state_dropdown.observe(update_city_options, names='value')
city_dropdown.observe(display_weather_data, names='value')

display(state_dropdown)
display(city_dropdown)
display(output)
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['Temperature (°C)'], data['Humidity (%)'])
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")
plt.title("Temperature vs. Humidity")
plt.show()
```

```python
import matplotlib.pyplot as plt

avg_temp_by_state = data.groupby('State')['Temperature (°C)'].mean()
plt.bar(avg_temp_by_state.index, avg_temp_by_state.values)
plt.xlabel("State")
plt.ylabel("Average Temperature (°C)")
plt.title("Average Temperature by State")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
