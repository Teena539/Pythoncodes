```python
# prompt: Using dataframe data: give python code for a/b testing BASED ON MY DATASET

import pandas as pd
from scipy.stats import ttest_ind

# Assuming 'data' is your DataFrame
# Replace 'A' and 'B' with the actual column names representing your A/B groups
# Replace 'conversion' with the actual column name representing your conversion metric (e.g., purchase, click, etc.)

def perform_ab_test(data, group_column, conversion_column):
    """
    Performs an A/B test using a t-test for two independent samples.

    Args:
        data (pd.DataFrame): The DataFrame containing the A/B test data.
        group_column (str): The name of the column that identifies the group (A or B).
        conversion_column (str): The name of the column that represents the conversion metric.

    Returns:
        tuple: A tuple containing the t-statistic, p-value, and a string summarizing the results.
    """
    # Separate data into A and B groups
    group_a = data[data[group_column] == 'A'][conversion_column]
    group_b = data[data[group_column] == 'B'][conversion_column]

    # Perform t-test
    t_statistic, p_value = ttest_ind(group_a, group_b)

    # Summarize results
    alpha = 0.05
    if p_value < alpha:
        result_summary = f"Significant difference between groups (p-value = {p_value:.3f})."
    else:
        result_summary = f"No significant difference between groups (p-value = {p_value:.3f})."

    return t_statistic, p_value, result_summary

# Example usage (assuming you have a 'Group' column indicating A or B and 'Purchase' is the conversion)
# If your data does not have a 'Group' Column, you will need to create one that labels each user as being in either group A or B

#Create a column called 'Group' to demonstrate
import numpy as np
data['Group'] = np.random.choice(['A', 'B'], size=len(data))

t_stat, p_val, summary = perform_ab_test(data, 'Group', 'Final_Price(Rs.)') #Change 'Final_Price(Rs.)' to the column representing your conversion
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
summary
```
