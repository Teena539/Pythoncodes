```python
import pandas as pd
data = pd.read_csv('/content/dataset.csv', encoding='latin-1')
```

```python
# TASK 02 - CHECKING MISSING VALUES & IF THERE ARE SOME DUBLICATE ROWS OR NOT

# Check for missing values in the dataframe
print(data.isnull().sum())

# Check for duplicate rows in the dataframe
print(f"\nNumber of duplicate rows: {data.duplicated().sum()}")
```

```python
#  TASK 03 - PLOTTING THE DISTRIBUTION OF VARIOUS DISCRETE FEATURES ON SEASON/HOLIDAY/WEEKEND/WEATHERCODE

import matplotlib.pyplot as plt
import seaborn as sns

# List of discrete features to plot
discrete_features = ['is_holiday', 'is_weekend', 'season', 'weather_code']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
axes = axes.flatten()  # Flatten the axes array for easier iteration

# Iterate through features and create plots
for i, feature in enumerate(discrete_features):
    sns.countplot(x=feature, data=data, ax=axes[i])
    axes[i].set_title(f'Distribution of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Count')

# Adjust layout and display plot
plt.tight_layout()
plt.show()
```

```python
#  #TASK 04 - LOOKING AT THE DATA TYPES OF EACH VARIABLE, TRANSFORM TIMESTAMP IN TYPE & SET IT AS INDEX# prompt: Using dataframe Set: LOOKING AT THE DATA TYPES OF EACH VARIABLE, TRANSFORM TIMESTAMP IN TYPE & SET IT AS INDEX# Check if 'timestamp' is already the indexif 'timestamp' not in Set.columns:     # If 'timestamp' is not a column, it's likely the index# Reset the index to make 'timestamp' a column againSet = Set.reset_index() # Convert 'timestamp' column to datetime objectsSet['timestamp'] = pd.to_datetime(Set['timestamp']) # Set the 'timestamp' column as the index of the DataFrameSet = Set.set_index('timestamp')print(Set)(i need this code for my data)

# Assuming 'Set' is your DataFrame (replace with your actual DataFrame name if different)
# and it's already loaded into the Colab environment.

# Check if 'timestamp' is already the index
if 'timestamp' not in data.columns: # Changed 'Set' to 'data'
    # If 'timestamp' is not a column, it's likely the index
    # Reset the index to make 'timestamp' a column again
    data = data.reset_index() # Changed 'Set' to 'data'

# Convert 'timestamp' column to datetime objects
data['timestamp'] = pd.to_datetime(data['timestamp']) # Changed 'Set' to 'data'

# Set the 'timestamp' column as the index of the DataFrame
data = data.set_index('timestamp') # Changed 'Set' to 'data'

data # Changed 'Set' to 'data'
```

```python
#  TASK 05 - MAKING FEATURE ENGINEERING. EXTRACTING NEW COLUMNS (DAY OF WEEK, DAY OF MONTH, HOUR, MONTH, SEASON, YEAR ETC.

# Extract year, month, day, hour, dayofweek from the timestamp index
data['year'] = data.index.year
data['month'] = data.index.month
data['day'] = data.index.day
data['hour'] = data.index.hour
data['dayofweek'] = data.index.dayofweek

# Define a function to map month to season
def map_month_to_season(month):
    if 3 <= month <= 5:
        return 'Spring'
    elif 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Autumn'
    else:
        return 'Winter'

# Apply the function to create the season column
data['season_name'] = data['month'].apply(map_month_to_season)

# Display first 5 rows of the updated dataframe
data.head()
```

```python
#  TASK 06 - VISUALIZING THE CORRELATION WITH A HEATMAP

import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix
corr_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Bike Sharing Data')
plt.show()
```

```python
#  TASK 07 - VISUALIZING THE CORRELATION OF THE TARGET VARIABLE & THE OTHER FEATURES WITH BARPLOT

# Plotting the correlation of the target variable with other features using barplot

plt.figure(figsize=(10, 6))
# Exclude non-numeric columns from correlation calculation
numeric_data = data.select_dtypes(include=['number'])  # Select only numeric columns
numeric_data.corr()['cnt'].sort_values(ascending=False).drop('cnt').plot(kind='bar') # Calculate correlation on numeric data only
plt.title('Correlation of Features with Target Variable (cnt)')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# TASK 08 - PLOTTING BIKE SHARES OVER TIME BY USING LINEPLOT

# Plotting bike shares over time using a line plot

plt.figure(figsize=(12, 6))
sns.lineplot(x=data.index, y='cnt', data=data)  # Use the timestamp index for the x-axis
plt.title('Bike Shares Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Count of Bike Shares')
plt.show()
```

```python
#  TASK 09 - PLOTTING BIKE SHARES BY MONTHS & YEAR_OF_MONTH (USE LINEPLOT, POINTPOT, BARPLOT)

# Plotting bike shares by month and year using different plot types

# Line plot of bike shares by month
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='cnt', data=data)
plt.title('Bike Shares by Month')
plt.xlabel('Month')
plt.ylabel('Count of Bike Shares')
plt.show()

# Point plot of bike shares by month and year
plt.figure(figsize=(12, 6))
sns.pointplot(x='month', y='cnt', hue='year', data=data)
plt.title('Bike Shares by Month and Year')
plt.xlabel('Month')
plt.ylabel('Count of Bike Shares')
plt.show()

# Bar plot of average bike shares by month and year
plt.figure(figsize=(12, 6))
sns.barplot(x='month', y='cnt', hue='year', data=data, ci=None) # ci=None removes confidence intervals
plt.title('Average Bike Shares by Month and Year')
plt.xlabel('Month')
plt.ylabel('Average Count of Bike Shares')
plt.show()
```

```python
#  TASK 10 - PLOTTING BIKE SHARES BY HOURS ON (HOLIDAYS, WEEKEND, SEASON)

# Plotting bike shares by hour for holidays, weekends, and seasons

plt.figure(figsize=(12, 6))
sns.pointplot(x='hour', y='cnt', hue='is_holiday', data=data)
plt.title('Bike Shares by Hour (Holidays)')
plt.xlabel('Hour')
plt.ylabel('Count of Bike Shares')
plt.show()

plt.figure(figsize=(12, 6))
sns.pointplot(x='hour', y='cnt', hue='is_weekend', data=data)
plt.title('Bike Shares by Hour (Weekends)')
plt.xlabel('Hour')
plt.ylabel('Count of Bike Shares')
plt.show()

plt.figure(figsize=(12, 6))
sns.pointplot(x='hour', y='cnt', hue='season', data=data)
plt.title('Bike Shares by Hour (Seasons)')
plt.xlabel('Hour')
plt.ylabel('Count of Bike Shares')
plt.show()

plt.figure(figsize=(12, 6))
sns.pointplot(x='hour', y='cnt', hue='season_name', data=data)
plt.title('Bike Shares by Hour (Seasons - Season Names)')
plt.xlabel('Hour')
plt.ylabel('Count of Bike Shares')
plt.show()
```

```python
#  TASK 11 - PLOTTING BIKE SHARES BY DAY OF WEEK

# Plotting bike shares by day of week

plt.figure(figsize=(10, 6))
sns.boxplot(x='dayofweek', y='cnt', data=data)
plt.title('Bike Shares by Day of Week')
plt.xlabel('Day of Week (0=Monday, 6=Sunday)')
plt.ylabel('Count of Bike Shares')
plt.show()
```

```python
#  TASK 12 - PLOTTING BIKE SHARES BY DAY OF MONTH

# Plotting bike shares by day of month

plt.figure(figsize=(12, 6))
sns.boxplot(x='day', y='cnt', data=data)
plt.title('Bike Shares by Day of Month')
plt.xlabel('Day of Month')
plt.ylabel('Count of Bike Shares')
plt.show()
```

```python
# TASK 13 - PLOTTING BIKE SHARES BY YEAR & PLOTTING BIKE SHARES ON HOLIDAYS BY SEASON

import matplotlib.pyplot as plt
import seaborn as sns

# Plotting bike shares by year
plt.figure(figsize=(12, 6))
sns.pointplot(x='year', y='cnt', data=data)
plt.title('Bike Shares by Year')
plt.xlabel('Year')
plt.ylabel('Count of Bike Shares')
plt.show()


# Plotting bike shares on holidays by season
plt.figure(figsize=(12, 6))
sns.pointplot(x='season', y='cnt', hue='is_holiday', data=data)
plt.title('Bike Shares on Holidays by Season')
plt.xlabel('Season')
plt.ylabel('Count of Bike Shares')
plt.show()
```

```python
# TASK 14 - VISUALIZING THE DISTRIBUTION OF BIKE SHARES BY WEEKDAY-WEEKEND WITH USING PIECHART & BARPLOT

# Calculate the total bike shares for weekdays and weekends
weekday_shares = data[data['is_weekend'] == 0]['cnt'].sum()
weekend_shares = data[data['is_weekend'] == 1]['cnt'].sum()

# Create a pie chart
labels = ['Weekdays', 'Weekends']
sizes = [weekday_shares, weekend_shares]
colors = ['skyblue', 'lightcoral']
explode = (0.1, 0)  # explode the 1st slice

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Distribution of Bike Shares by Weekday/Weekend')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()


# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=labels, y=[weekday_shares, weekend_shares])
plt.title('Bike Shares by Weekday/Weekend')
plt.xlabel('Day Type')
plt.ylabel('Total Bike Shares')
plt.show()
```

```python
# TASK 15 - PLOTTING THE DISTRIBUTION OF WEATHER CODE BY SEASON

# Plotting the distribution of weather code by season
plt.figure(figsize=(10, 6))
sns.countplot(x='weather_code', hue='season_name', data=data)
plt.title('Distribution of Weather Code by Season')
plt.xlabel('Weather Code')
plt.ylabel('Count')
plt.show()
```
