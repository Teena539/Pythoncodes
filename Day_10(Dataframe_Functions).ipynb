```
import numpy as np
import pandas as pd
```

```
data = pd.read_csv('/content/student-lifestyle-dataset/student_lifestyle_dataset.csv')
```

```
data
```

```
#  data.loc[]

# Display the first row of the dataframe.
data.loc[0]
```

loc is primarily **label-based indexing**. This means that you **select rows and columns using their labels** (row and column names)

```
# data.iloc[]

# Display the second row of the dataframe (index 1).
data.iloc[1]
```

iloc is a **integer position-based indexing.** This means that you select rows and columns using their **integer positions** (starting from 0).

```
#  Axis0

# Calculate the sum of each column (axis=0).
column_sums = data.sum(axis=0)
column_sums
```

**axis=0** refers to the **rows of the DataFrame.**

```
# Axis1

# Calculate the sum of each row (axis=1) for numeric columns only.
row_sums = data.select_dtypes(include=np.number).sum(axis=1)
row_sums
```

**axis=1** refers to the **columns of the DataFrame**
