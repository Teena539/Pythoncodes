```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame

# 1. Select numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# 2. Melt the DataFrame to create a long-form dataset
melted_df = pd.melt(df, value_vars=numerical_cols, var_name='Feature', value_name='Value')

# 3. Create a single box plot for all features
plt.figure(figsize=(12, 8))  # Adjust figure size if needed
sns.boxplot(x='Feature', y='Value', data=melted_df)
plt.title('Box Plot of All Numerical Features')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.show()
```
