```
# prompt: Using dataframe data: generate python code using the concepts like nominal,ordinal,measures of data,variance,standard deviation,range,skewness,outliers based on the above dataset

import pandas as pd
import numpy as np

# Assuming your data is in a pandas DataFrame called 'data'

# Nominal data analysis (categorical variables)
# Example: Device Model, Operating System, Gender
# We can analyze the frequency distribution of each category
device_model_counts = data['Device Model'].value_counts()
operating_system_counts = data['Operating System'].value_counts()
gender_counts = data['Gender'].value_counts()

print("Device Model Counts:\n", device_model_counts)
print("\nOperating System Counts:\n", operating_system_counts)
print("\nGender Counts:\n", gender_counts)

# Ordinal data analysis (categorical variables with inherent order)
# Example: User Behavior Class (assuming the numbers represent a ranking)
# We can calculate the frequency of each behavior class
behavior_class_counts = data['User Behavior Class'].value_counts()
print("\nUser Behavior Class Counts:\n", behavior_class_counts)

# Measures of data, variance, standard deviation, range, skewness for numerical variables

numerical_columns = ['App Usage Time (min/day)', 'Screen On Time (hours/day)',
                     'Battery Drain (mAh/day)', 'Number of Apps Installed',
                     'Data Usage (MB/day)', 'Age']


for col in numerical_columns:
    print(f"\nAnalysis for {col}:")

    # Measures
    mean = data[col].mean()
    median = data[col].median()
    mode = data[col].mode()[0]  # Get the first mode value

    print(f"  Mean: {mean}")
    print(f"  Median: {median}")
    print(f"  Mode: {mode}")

    # Variance and Standard Deviation
    variance = np.var(data[col])
    std_dev = np.std(data[col])
    print(f"  Variance: {variance}")
    print(f"  Standard Deviation: {std_dev}")

    # Range
    data_range = data[col].max() - data[col].min()
    print(f"  Range: {data_range}")

    # Skewness
    skewness = data[col].skew()
    print(f"  Skewness: {skewness}")

    # Outliers (using IQR method)
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]

    print(f"  Outliers:\n{outliers}")
```
