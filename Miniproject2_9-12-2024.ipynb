 **TASK 12 :-** Create a new column for the age of the car using the registration field

```python
# prompt: Create a new column for the age of the car using the registration field.

from datetime import datetime

current_year = datetime.now().year

# Function to calculate car age
def calculate_car_age(registration_date):
    try:
        registration_year = registration_date.year
        age = current_year - registration_year
        return age
    except:
        return None

# Apply the function to create the 'car_age' column
df['car_age'] = df['registration'].apply(calculate_car_age)

df
```

```python
# prompt: Generate a price_per_km column to assess cost efficiency.

# Calculate price per kilometer
df['price_per_km'] = df['price'] / df['km']

# Display the DataFrame with the new column
df
```

**TASK 14 :-** Create categories for cars based on price (e.g., low, mid, high)

```python
# prompt: Create categories for cars based on price (e.g., low, mid, high).

# Define price categories based on quantiles
price_quantiles = df['price'].quantile([0.33, 0.66])

def categorize_price(price):
    if price <= price_quantiles[0.33]:
        return 'low'
    elif price <= price_quantiles[0.66]:
        return 'mid'
    else:
        return 'high'

df['price_category'] = df['price'].apply(categorize_price)

print(df[['price', 'price_category']])
```

**TASK 19 :-** Group the vehicles based on attributes like price, km, and age

```python
# prompt: Group the vehicles based on attributes like price, km, and age.

# Group vehicles based on price, km, and age categories
# Define bins for price, km, and age
price_bins = [0, 10000, 20000, 30000, float('inf')]
km_bins = [0, 50000, 100000, 150000, float('inf')]
age_bins = [0, 2, 5, 10, float('inf')]

# Create labels for the bins
price_labels = ['Low', 'Mid', 'High', 'Very High']
km_labels = ['Low', 'Mid', 'High', 'Very High']
age_labels = ['New', 'Used', 'Older', 'Very Old']

# Cut the data into categories based on the bins
df['price_group'] = pd.cut(df['price'], bins=price_bins, labels=price_labels, right=False)
df['km_group'] = pd.cut(df['km'], bins=km_bins, labels=km_labels, right=False)
df['age_group'] = pd.cut(df['car_age'], bins=age_bins, labels=age_labels, right=False)

# Group the data based on the created categories
grouped_vehicles = df.groupby(['price_group', 'km_group', 'age_group'])

# Display the grouped data
print("\nGrouped Vehicles:")
print(grouped_vehicles.size().unstack(fill_value=0))
```
