```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec
from sklearn.metrics import confusion_matrix
data = pd.read_csv('/content/sentiment-analysis/sentiment_analysis.csv')
data
```

```

# Filter data for positive and negative sentiments only
filtered_data = data[data['sentiment'].isin(['positive', 'negative'])]

# Assuming you have already trained a model and obtained predicted sentiments
# For demonstration, I'm using random predictions. Replace this with your predictions.
import numpy as np
y_true = filtered_data['sentiment']
y_pred = np.random.choice(['positive', 'negative'], size=len(y_true))


cm = confusion_matrix(y_true, y_pred, labels=['positive', 'negative'])

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['positive', 'negative'], yticklabels=['positive', 'negative'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```
# Example true and predicted labels (replace with your actual data)
y_true = np.random.choice(['positive', 'negative', 'neutral'], size=len(data))
y_pred = np.random.choice(['positive', 'negative', 'neutral'], size=len(data))


cm = confusion_matrix(y_true, y_pred, labels=['positive', 'negative', 'neutral'])

# Create a heatmap of the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=['positive', 'negative', 'neutral'],
            yticklabels=['positive', 'negative', 'neutral'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

# **Business Problem**

The business problem revolves around understanding customer sentiment expressed in text data. Businesses often need to analyze reviews, feedback, or social media posts to gauge public opinion about their products or services.

# **Business Objective**

The objective is to build a sentiment analysis model that can accurately classify text into different sentiment categories (e.g., positive, negative, neutral). This allows businesses to:

Automate the analysis of large volumes of text data.
Identify areas for improvement based on customer feedback.
Track changes in sentiment over time.
Make data-driven decisions about product development or marketing strategies.

# **Business Constraints**

**Data Availability:**

 Having enough labeled data to train a robust and accurate model.

**Model Complexity:**

Balancing model complexity with performance and interpretability.

**Ambiguity in Language: **bold text**

Dealing with the nuances of human language, such as sarcasm or irony, which can be difficult for models to understand.

**Real-time Performance:**
 If the analysis needs to be done in real-time, there may be constraints on the model's processing speed.
