## **ASSIGNMENT2:-** Data Cleaning

### **Task1:-** If there are duplicates, remove them

```
# Task1:- If there are duplicates, remove them.

# Drop duplicate rows
data = data.drop_duplicates()

# Verify if duplicates are removed
duplicate_rows_after_removal = data.duplicated().sum()
print("\nNumber of duplicate rows after removal:", duplicate_rows_after_removal)
```

### **Task2:-** Handle missing values appropriately (drop or impute them).

```
# Task2:-	Handle missing values appropriately (drop or impute them).

# Handle missing values (example: imputation with mean for numerical columns)
for col in data.select_dtypes(include=['number']):
  if data[col].isnull().any():
    data[col] = data[col].fillna(data[col].mean())

# Alternative: Drop rows with any missing values
data = data.dropna()

#Check if any missing values are remaining
missing_values_after_handling = data.isnull().sum()
print("\nMissing Values after handling:\n", missing_values_after_handling)
```

## **ASSIGNMENT4:-** Timestamp Transformation

### **Task1:-**	Convert the Order Date column to a datetime format

```
# Task1:-	Convert the Order Date column to a datetime format.

# Convert 'Order Date' column to datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'])
data
```

### **Task2:-** Create new columns : Day of the week,Day of the month,Month,Year,Season(Winter,Spring,Summer,Fall)

```
#Task 2:-	Create new columns:
# o	Day of the week.
# o	Day of the month.
# o	Month.
# o	Year.
# o	Season (Winter, Spring, Summer, Fall).

# Create new columns for day of the week, day of the month, month, year, and season
data['Day of the week'] = data['Order Date'].dt.day_name()
data['Day of the month'] = data['Order Date'].dt.day
data['Month'] = data['Order Date'].dt.month_name()
data['Year'] = data['Order Date'].dt.year

# Define a function to determine the season based on the month
def get_season(month):
  if month in [12, 1, 2]:
    return 'Winter'
  elif month in [3, 4, 5]:
    return 'Spring'
  elif month in [6, 7, 8]:
    return 'Summer'
  else:
    return 'Fall'

# Apply the function to create the 'Season' column
data['Season'] = data['Month'].apply(lambda x: get_season(pd.to_datetime(x, format='%B').month))

data
```

### **Task3:-** Set the Order Date column as the DataFrame index

```
# prompt: Task3:- Set the Order Date column as the DataFrame index

# Set the 'Order Date' column as the DataFrame index
data = data.set_index('Order Date')
data
```
