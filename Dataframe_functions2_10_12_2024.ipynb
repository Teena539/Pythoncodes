# **Task1:-**

## **drop=True**

```
data = data.reset_index(drop=True)
data
```

### **drop= True**
### The old index is discarded and a new integer index is created.

```
data1 = data.reset_index(drop=False)
data1
```

###**drop=False**

###The old index is retained as a column in the DataFrame

```
# Convert 'InvoiceDate' to datetime objects if it's not already
data['Purchase_Date'] = pd.to_datetime(data['Purchase_Date'], format='%d-%m-%Y', errors='coerce')
#Added format='%d-%m-%Y' to specify the correct date format.
#errors='coerce' will handle invalid dates, turning them to NaT (Not a Time)


# Extract month and year
data['month'] = data['Purchase_Date'].dt.month
data['year'] = data['Purchase_Date'].dt.year
```

```
data
```

 If we use data.to_csv('dataset.csv', index=False), it downloads an  updated version of the dataset
