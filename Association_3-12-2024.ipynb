```
import pandas as pd
data = pd.read_csv('/content/Policedataset.csv', encoding='latin-1')
```

```
# prompt: Using dataframe data: association for using only two coulmns here take 2 columns as id and age

import pandas as pd

# Assuming 'data' is your DataFrame
# Create a new DataFrame with only the 'id' and 'age' columns
new_data = data[['id', 'age']]

# Display the first few rows of the new DataFrame
print(new_data.head())
```

```
# prompt: Using dataframe data: dependence

import pandas as pd

# Assuming 'data' is your DataFrame

# Calculate the correlation between age and signs_of_mental_illness
correlation = data['age'].corr(data['signs_of_mental_illness'].astype(int)) # Convert boolean to integer for correlation

print(f"Correlation between age and signs_of_mental_illness: {correlation}")

# Group data by gender and calculate the mean age for each gender
mean_age_by_gender = data.groupby('gender')['age'].mean()

print("\nMean age by gender:")
print(mean_age_by_gender)

# Create a cross-tabulation of gender and manner_of_death
gender_manner_of_death_crosstab = pd.crosstab(data['gender'], data['manner_of_death'])

print("\nCross-tabulation of gender and manner_of_death:")
gender_manner_of_death_crosstab
```

```
# prompt: Using dataframe data: correlation

# Calculate the correlation between age and id
correlation = data['age'].corr(data['id'])

# Print the correlation
print(f"The correlation between age and id is: {correlation}")
```

```
# prompt: Using dataframe data: casuation

# Convert the 'date' column to datetime objects if it's not already
# and handle potential errors gracefully
import pandas as pd

try:
    data['date'] = pd.to_datetime(data['date'], errors='coerce')
except (ValueError, TypeError):
    print("Error converting 'date' column to datetime. Check the data format.")


# Calculate the correlation between 'age' and 'signs_of_mental_illness'
correlation = data['age'].corr(data['signs_of_mental_illness'].astype(int))

print(f"Correlation between age and signs_of_mental_illness: {correlation}")

# Group data by 'manner_of_death' and calculate the mean age for each group
mean_age_by_manner_of_death = data.groupby('manner_of_death')['age'].mean()

print(f"\nMean age by manner of death:\n{mean_age_by_manner_of_death}")
```

```
# Calculate the correlation matrix for numerical columns only.
correlation_matrix = data.select_dtypes(include=['number']).corr()

# Display the correlation matrix.
correlation_matrix
```

```
# prompt: Using dataframe data: 10 different graphs for matplot lib

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is your pandas DataFrame

# 1. Bar plot of manner of death
plt.figure(figsize=(8, 6))
sns.countplot(x='manner_of_death', data=data)
plt.title('Distribution of Manner of Death')
plt.show()

# 2. Histogram of age
plt.figure(figsize=(8, 6))
sns.histplot(data['age'], kde=True)
plt.title('Distribution of Age')
plt.show()

# 3. Countplot of gender
plt.figure(figsize=(6, 4))
sns.countplot(x='gender', data=data)
plt.title('Count of Gender')
plt.show()


# 4. Pie Chart of Race
race_counts = data['race'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Race')
plt.show()


# 5. Scatter plot age vs. gender
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='gender', data=data, hue='gender')
plt.title('Age vs. Gender')
plt.show()

# 6. Box plot of age by gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='age', data=data)
plt.title('Age Distribution by Gender')
plt.show()

# 7. Violin plot age by race
plt.figure(figsize=(10,6))
sns.violinplot(x='race', y='age', data=data)
plt.title('Age Distribution by Race')
plt.show()

# 8. Countplot of Threat Level
plt.figure(figsize=(8, 6))
sns.countplot(x='threat_level', data=data)
plt.title('Distribution of Threat Levels')
plt.show()


# 9. Bar plot of 'armed' categories
plt.figure(figsize=(12, 6))
armed_counts = data['armed'].value_counts().head(10) # Show top 10 armed categories
sns.barplot(x=armed_counts.index, y=armed_counts.values)
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Armed Categories')
plt.show()

#10. Line plot of Number of deaths per year
data['year'] = data['date'].dt.year #Extract year from date
yearly_deaths = data.groupby('year').size()
plt.figure(figsize=(8, 6))
plt.plot(yearly_deaths.index, yearly_deaths.values)
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.title('Number of Deaths Over Time')
plt.show()
```

```
# prompt: Using dataframe data: 10 different graphs for seaborn library

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your Pandas DataFrame

# 1. Histogram of age
sns.histplot(data=data, x="age")
plt.title("Distribution of Age")
plt.show()

# 2. Countplot of manner of death
sns.countplot(data=data, x="manner_of_death")
plt.title("Count of Manner of Death")
plt.show()


# 3. Boxplot of age by gender
sns.boxplot(data=data, x="gender", y="age")
plt.title("Age Distribution by Gender")
plt.show()


# 4. Violinplot of age by race
sns.violinplot(data=data, x="race", y="age")
plt.title("Age Distribution by Race")
plt.show()

# 5. Scatterplot of age vs. year
sns.scatterplot(data=data, x="year", y="age")
plt.title("Age vs. Year")
plt.show()

# 6. Bar plot of threat level
sns.countplot(x="threat_level", data=data)
plt.title("Threat Level Counts")
plt.show()


# 7. Heatmap of correlation between numerical features (assuming you have more numerical data)
numeric_data = data.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()


# 8. Pairplot for a subset of numerical columns.
sns.pairplot(data[["age", "year"]])
plt.show()

# 9. Jointplot for age and year
sns.jointplot(data=data, x="age", y="year", kind="kde")
plt.show()

#10.  KDE Plot for Age
sns.kdeplot(data=data, x="age", shade=True)
plt.title("KDE Plot of Age")
plt.show()
```
